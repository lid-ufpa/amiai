# Selection and Pre-Processing
In this present notebook, we selected relevant columns, cleaned our database with regex and own functions and pre-processed texts with tokenization, removing stopwords and lemmatization.

✈️ We will use natural language processing techniques to apply machine learning algorithms to the data in next notebook.

## Selection

```python
import pandas as pd
import sys
import os
sys.path.append(os.path.abspath(os.path.join('..')))
from src import cleaning
from src import preprocessing
```

```python
df = pd.read_csv("../data/00_amiai_raw_data.csv")
df.head()
```

```python
df = df.rename(columns={'O que te motivou a estudar Inteligência Artificial? (até 15 linhas)': 'background',
                        'Qual é o seu perfil de estudante?': 'student_profile',
                        'Carimbo de data/hora': 'date_time'})
df.head()
```

## Cleaning


```python
#quickly clean
df['background'] = cleaning.clean_text(df['background'])
df['background'] = df['background'].apply(lambda x: x.lower() if isinstance(x, str) else x)
df.head()
```

## Pre-Processing

```python
df['tokens'] = df['background'].apply(preprocessing.tokenize)
df['tokens'] = df['tokens'].apply(preprocessing.remove_stopwords)
df['stems'] = df['tokens'].apply(preprocessing.stemming)
df['lemmas'] = df['background'].apply(preprocessing.lemmatization)

df.head()
```

```python
df.to_csv('../data/01_amiai_preprocessed_data.csv')
```
